```python
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X,y=mglearn.datasets.make_wave(n_samples=60)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
lr=LinearRegression().fit(X_train,y_train)
print("lr.coef_:{}".format(lr.coef_))
print("lr.intercept_:{}".format(lr.intercept_))
print("Training set score:{:.2f}".format(lr.score(X_train,y_train)))
```

```python
import mglearn
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.load_extended_boston()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge().fit(X_train,y_train)
print("Training set score:",format(ridge.score(X_train,y_train)))
print("Test set score:",format(ridge.score(X_test,y_test)))
```

```python
from sklearn.linear_model import Ridge

alphas = [0.001, 0.01, 0.1, 1, 10, 100]
for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X_train, y_train)
    print(f"Alpha: {a:.3f}")
    print(f"Training score: {ridge.score(X_train, y_train):.3f}")
    print(f"Test score: {ridge.score(X_test, y_test):.3f}")
    print("-" * 30)
```

```python
import mglearn
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.load_extended_boston()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

lasso=Lasso().fit(X_train,y_train)
print("Training set score:",format(lasso.score(X_train,y_train)))
print("Test set score:",format(lasso.score(X_test,y_test)))
```

```python
alphas=[0.01,0.1,1,0.2,0.3,0.4,10,100]
for i in alphas:
      lasso2=Lasso(alpha=i ,max_iter=10000).fit(X_train,y_train)
      print("Training set score for alpha {}: {:.2f}".format(i,lasso2.score(X_train,y_train)))
      print("Test set score for alpha {}: {:.2f}".format(i,lasso2.score(X_test,y_test)))
      print("*"*30)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


%pip install load_breast_cancer
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=66)
training_accuracy=[]
test_accuracy=[]
neighbors_settings=range(1,11)
for n_neighbors in neighbors_settings:
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    training_accuracy.append(clf.score(X_train,y_train))
    test_accuracy.append(clf.score(X_test,y_test))
```
