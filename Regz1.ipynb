This Session Elaborates Linear Regression in Relation to Machine Learning

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([5, 12, 20, 30, 40, 50]).reshape((-1, 1))
y = np.array([10, 20, 18, 28, 25, 67])
model = LinearRegression()
model.fit(x, y)
print('The Y-Intercept:', model.intercept_)
print('The Slope:', model.coef_)
```

```python
y_hat=model.coef_*x+model.intercept_
plt.scatter(x,y)
```

```python
accuracy=model.score(x,y)
print("The Model accuracy is", accuracy)
```

```python
unknown_values=np.array([23,35,78]).reshape(-1,1)
result=model.predict(unknown_values)
print("The model Predictions are",result)
```

```python
plt.plot(x,y_hat,color='r')
plt.xlabel("X")
plt.ylabel("Y")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mglearn
x,y=mglearn.datasets.make_forge()
data=pd.DataFrame(np.c_[x,y],columns=['feature 1','Feature 2','Class'])
data
```

```python
mglearn.discrete_scatter(x[:,0],x[:,1],y)
plt.legend(["Class 0","Class 1"])
plt.xlabel("First Feature")
plt.ylabel("Second Feature")
plt.show()
```

KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)

print("Actual Class Labels:", y_test)
print("Test set predictions:", model.predict(x_test))
print("Model Accuracy:",model.score(x_test,y_test))
```
