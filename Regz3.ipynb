```python
from sklearn.datasets import load_iris
Iris_dataset = load_iris()
Iris_dataset
```

```python
print("Keys of Iris_dataset\n{}".format(Iris_dataset.keys()))
```

```python
print(Iris_dataset['DESCR'][:193]+"\n....")
```

```python
print("First five columns of data: \n{}".format(Iris_dataset['data'][:5]))
```

```python
import pandas as pd
import numpy as np

Iris_df=pd.DataFrame(data=np.c_[Iris_dataset['data'], Iris_dataset['target']], columns=Iris_dataset['feature_names']+['target'])
Iris_df
```

```python
Iris_df.isnull().sum()
```

```python
grr=pd.plotting.scatter_matrix(Iris_df, c=Iris_df['target'], figsize=(10,10), marker='o', hist_kwds={'bins':20}, s=60, alpha=.8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Iris_dataset['data'], Iris_dataset['target'], random_state=0, test_size=0.3)

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=17)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
y_test
```

```python
accuracy=model.score(X_test, y_test)
accuracy
```

```python
x_new=np.array([[4.5,2.3,6.9,2.5]])
prediction=model.predict(x_new)
prediction
print("prediction target name is {}".format(Iris_dataset['target_names'][prediction]))
```
