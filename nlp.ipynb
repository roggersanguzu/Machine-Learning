```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
```

```python
nlp=MLPClassifier(random_state=42)
nlp.fit(X_train,y_train)

print("The train Data result is",nlp.score(X_train,y_train))
print("The Test Data result is",nlp.score(X_test,y_test))
```

```python
nlp2=MLPClassifier(random_state=0,hidden_layer_sizes=[50,100])
nlp2.fit(X_train,y_train)

print("The train Data result is",nlp2.score(X_train,y_train))
print("The Test Data result is",nlp2.score(X_test,y_test))
```

```python
nlp3=MLPClassifier(random_state=0,hidden_layer_sizes=[50,100],activation="tanh")
nlp3.fit(X_train,y_train)
print("The train Data result is",nlp3.score(X_train,y_train))
print("The Test Data result is",nlp3.score(X_test,y_test))
```

```python
nlp4=MLPClassifier(random_state=0,hidden_layer_sizes=[50,100],activation="tanh",alpha=0.1)
nlp4.fit(X_train,y_train)
print("The train Data result is",nlp4.score(X_train,y_train))
print("The Test Data result is",nlp4.score(X_test,y_test))
```
